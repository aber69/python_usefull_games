```python
aaa_line = np.arange(1, 1000)
aaa_line = np.vstack([aaa_line, aaa_line[::-1], aaa_line])
aaa_line.shape
```

```python
plt.figure(figsize=(38.71, 2), dpi=100, frameon=False)
plt.imshow(aaa_line, aspect='auto', norm=matplotlib.colors.LogNorm(.9), cmap='turbo')
```

```python
aaa_line_norm = matplotlib.colors.LogNorm(.9)(aaa_line)
aaa_line_norm.shape
```

```python
plt.figure(figsize=(38.71, 2), dpi=100, frameon=False)
plt.imshow(aaa_line_norm, aspect='auto', cmap='turbo')
```

```python
aaa_line_colored = matplotlib.cm._cmap_registry['turbo'](aaa_line_norm)
aaa_line_colored.shape
```

```python
plt.figure(figsize=(38.71, 2), dpi=100, frameon=False)
plt.imshow(aaa_line_colored, aspect='auto')
```

## 2D Case

```python
aaa = np.arange(0, 50*30).reshape(30,50)
np.min(aaa), np.max(aaa)
```

```python
plt.figure(figsize=(15, 6), dpi=100, frameon=False)
plt.imshow(aaa, aspect='auto', norm=matplotlib.colors.Normalize(), cmap='turbo');
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

```python
plt.figure(figsize=(15, 6), dpi=100, frameon=False)
plt.imshow(matplotlib.colors.Normalize()(aaa), aspect='auto', cmap='turbo')
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

```python
plt.figure(figsize=(15, 6), dpi=100, frameon=False)
plt.imshow(matplotlib.cm._cmap_registry['turbo'](matplotlib.colors.Normalize()(aaa)), aspect='auto', cmap='gray')
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

## 2D Case - with exeptions and clipping

```python
bbb = np.cos(aaa/230)+0.99
plt.figure(figsize=(20,7), frameon=False)
plt.imshow(bbb, aspect='auto', norm=matplotlib.colors.LogNorm(.1), cmap='turbo');
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

```python
## !!! Attention: totally diffent map (in low part of values it more explicit)

plt.figure(figsize=(20,7), frameon=False)
plt.imshow(matplotlib.colors.LogNorm(.1)(bbb), aspect='auto', cmap='turbo');
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

```python
# !!! Now it is again correct (after external clipping)!!!
bbb1 = bbb.copy()
bbb1[np.logical_and(bbb1 > 0, bbb1 < 0.1)]=0.1 ## external clipping
plt.figure(figsize=(20,7), frameon=False)
plt.imshow(matplotlib.colors.LogNorm(.1)(bbb1), aspect='auto', cmap='turbo');
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

```python
# !!! Since bbb1 does not have values in range (0-0.1) we can use it without clipping in LogNorm
plt.figure(figsize=(20,7), frameon=False)
plt.imshow(matplotlib.colors.LogNorm()(bbb1), aspect='auto', cmap='turbo');
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```

```python
### Now apply colopmap expilictly (not inlfuence to cmap value)
plt.figure(figsize=(20,7), frameon=False)
plt.imshow(matplotlib.cm._cmap_registry['turbo'](matplotlib.colors.LogNorm()(bbb1)), aspect='auto', cmap='gray');
plt.colorbar(orientation='vertical', fraction=.04, aspect=100, pad=0.03)
```
